```python
import pandas as pd

ds = pd.read_excel('wanted_DS_cleansing_수정.xlsx')
ds.head(3)
```

```python
#preference 에 속하는 내용들을 하나의 리스트로 묶어주기

pref = ds['preference'].tolist()

pref = ' '.join(pref)
print(pref)
```

```python
tokenizer = Okt()
raw_pos_tagged = tokenizer.pos(pref, norm=True, stem=True) #pos tagging, norm은 정규화, stem은 어근화
print(raw_pos_tagged)
```

```python
word_cleaned = []

for word in raw_pos_tagged:
    if word[1] in ['Noun','Alpha']:
        if len(word[0]) != 1:
            word_cleaned.append(word[0])

print(word_cleaned)
```

```python
from collections import Counter

result = Counter(word_cleaned)
word_dic = dict(result)
print(word_dic)
```

```python
word_dic = sorted(word_dic.items(), key = lambda x:x[1], reverse = True)
print(word_dic)
```
